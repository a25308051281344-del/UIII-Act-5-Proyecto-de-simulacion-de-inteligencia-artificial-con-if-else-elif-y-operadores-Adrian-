```python
# =====================================
# IA Diseñador Virtual
# Adrian Hinojos Chavez
# =====================================

print("===================================")
print("   IA DISEÑADOR VIRTUAL")
print("===================================")

nombre = input("Nombre del cliente: ")

print("\nTipos de diseño")
print("1. Logo")
print("2. Página Web")
print("3. Infografía")
print("4. Cartel Publicitario")

opcion = input("Seleccione una opción: ")

presupuesto = int(input("Ingrese presupuesto: "))
urgente = input("¿Es urgente? (si/no): ")

print("\nAnalizando solicitud...")
print("-----------------------------------")

# Uso de IF

if opcion == "1":
    print("Proyecto seleccionado: Logo")

if opcion == "2":
    print("Proyecto seleccionado: Página Web")

if opcion == "3":
    print("Proyecto seleccionado: Infografía")

if opcion == "4":
    print("Proyecto seleccionado: Cartel Publicitario")


# Uso de IF ELSE

if urgente == "si":
    print("Prioridad Alta")
else:
    print("Prioridad Normal")


# Uso de ELIF y operadores

if presupuesto >= 10000:
    print("IA Recomienda: Diseño Premium")
    print("Incluye animaciones y elementos avanzados")

elif presupuesto >= 5000 and presupuesto < 10000:
    print("IA Recomienda: Diseño Profesional")
    print("Incluye recursos gráficos de alta calidad")

elif presupuesto >= 2000 and presupuesto < 5000:
    print("IA Recomienda: Diseño Estándar")
    print("Incluye diseño atractivo y funcional")

else:
    print("IA Recomienda: Diseño Básico")
    print("Ideal para proyectos escolares")


# Operadores relacionales

if presupuesto > 8000:
    print("Cliente con presupuesto alto")

if presupuesto < 2000:
    print("Cliente con presupuesto limitado")

if presupuesto == 5000:
    print("Presupuesto exacto para paquete profesional")


# Operadores lógicos

if urgente == "si" and presupuesto >= 5000:
    print("Resultado IA: Proyecto prioritario")

elif urgente == "si" or presupuesto >= 10000:
    print("Resultado IA: Atención especial")

else:
    print("Resultado IA: Flujo normal de trabajo")


print("\n===================================")
print("REPORTE DEL DISEÑADOR VIRTUAL")
print("Cliente:", nombre)
print("Presupuesto:", presupuesto)
print("Urgente:", urgente)
print("===================================")
```
